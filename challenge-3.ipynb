## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

# Problem 1
amount_spent_customers = orders.groupby(["CustomerID", "Country"])[["amount_spent"]].sum()
amount_spent_customers = amount_spent_customers.sort_values("amount_spent", ascending=False).reset_index()

# Problem 2
seventyfive_quantile = amount_spent_customers["amount_spent"].quantile(0.75)
ninetyfive_quantile = amount_spent_customers["amount_spent"].quantile(0.95)

# Problem 3
amount_spent_customers["Category"] = np.where(amount_spent_customers["amount_spent"] > ninetyfive_quantile, "VIP", np.where(amount_spent_customers["amount_spent"] > seventyfive_quantile, "Preferred", "-"))
amount_spent_customers
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
vip_customers = amount_spent_customers[amount_spent_customers["Category"] == "VIP"]
vip_customers
```

```python
vip_customers.groupby(["Country"])[["Category"]].count().sort_values("Category", ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
vip_and_preferred_customers = amount_spent_customers[(amount_spent_customers["Category"] == "VIP") | (amount_spent_customers["Category"] == "Preferred")]
vip_and_preferred_customers
```

```python
vip_and_preferred_customers.groupby(["Country"])[["Category"]].count().sort_values("Category", ascending=False)
```
