## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
orders = pd.read_csv("data/Orders/Orders.csv")
orders
```

```python
customers = orders[["CustomerID", "amount_spent"]].reset_index(drop=True).groupby(by="CustomerID").sum().sort_values(by= "amount_spent", ascending=False)
#customers
quantile_95 = customers["amount_spent"].quantile(0.95)
quantile_75 = customers["amount_spent"].quantile(0.75)
customers_types = ["VIP", "Prefered"]
conditions = [
    customers["amount_spent"] >  quantile_95, 
    (customers["amount_spent"] > quantile_75) & (customers["amount_spent"] < quantile_95)
]
customers["type"] = np.select(conditions, customers_types, default="Regular")
customers
```

```python
countries = orders[["CustomerID", "Country"]].drop_duplicates().reset_index(drop=True)
countries
```

```python
customers = customers.merge(countries, on="CustomerID", how="left")
customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
customers[customers["type"] == "VIP"].groupby(by = ["Country", "type"])["type"].count().sort_values(ascending = False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
customers[customers["type"] != "Regular"].groupby(by = ["Country"])["type"].count().sort_values(ascending = False)
```
