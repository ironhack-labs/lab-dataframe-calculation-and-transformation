## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

# orders.groupby("CustomerID").sum() # --> this is wrong as it sums all the columns for each customer and I want only the amount spent

# Group by 'CustomerID' and sum the 'amount_spent'
customer_spending = orders.groupby('CustomerID')['amount_spent'].sum()
display(customer_spending)
# Convert the Series into a DataFrame
customer_spending_df = customer_spending.reset_index()
customer_spending_df



```

##### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
# Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

# Calculate the 75th and 95th percentiles of the `amount_spent`
percentile_75 = customer_spending_df['amount_spent'].quantile(0.75)
percentile_95 = customer_spending_df['amount_spent'].quantile(0.95)

print(f"75th Percentile: {percentile_75}")
print(f"95th Percentile: {percentile_95}")

```

##### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?


```python
# Create a new column to categorize customers as 'Preferred' or 'VIP'
def categorize_customer(spent):
    if spent > percentile_95:
        return 'VIP'
    elif spent > percentile_75:
        return 'Preferred'
    else:
        return 'Regular'  # Optionally handle customers below 75th percentile

# Apply the categorization
customer_spending_df['Customer Category'] = customer_spending_df['amount_spent'].apply(categorize_customer)

# Preview the categorized DataFrame
display(customer_spending_df.head(100))
```

```python
display(customer_spending_df.columns)
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
# Step 1: Merge country information into the aggregated DataFrame
customer_spending_with_country = pd.merge(
    customer_spending_df,
    orders[['CustomerID', 'Country']].drop_duplicates(),
    on='CustomerID',
    how='left'
)
display(customer_spending_with_country)
```

```python
# Step 2: Filter for VIP customers
vip_customers = customer_spending_with_country[customer_spending_with_country['Customer Category'] == 'VIP']
display(vip_customers)
```

```python
# Step 3: Group by 'Country' and count number of VIPs
vip_count_by_country = vip_customers.groupby('Country').size().reset_index(name='VIP Count')
vip_count_by_country
```

```python
# Step 4: Identify country with the most VIPs
country_with_most_vips = vip_count_by_country.sort_values(by='VIP Count', ascending=False).head(1)
country_with_most_vips
print(f"The country {country_with_most_vips['Country']} has got the most VIP customers with {country_with_most_vips['VIP Count']} VIP Customers")
```

```python
print("The country {} has got the most VIP customers with {} VIP Customers.".format(country_with_most_vips['Country']
                                                                                    , country_with_most_vips['VIP Count']))
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# Step 1: Filter for VIP + Preferred customers (similar code than step 2 before but with 'pipe' as an 'or' logical function)
vip_preferred_customers = customer_spending_with_country[(customer_spending_with_country['Customer Category'] == 'VIP')
                                                          | (customer_spending_with_country['Customer Category'] == 'Preferred')]
display(vip_preferred_customers)
```

```python
# Step 2: Group by 'Country' and count number of VIPs
vip_preferred_count_by_country = vip_preferred_customers.groupby('Country').size().reset_index(name='VIP+Preferred Count')
vip_preferred_count_by_country
```

```python
# Step 3: Identify country with the most VIPs
country_with_most_vips_preferred = vip_preferred_count_by_country.sort_values(by='VIP+Preferred Count', ascending=False).head(1)
display(country_with_most_vips_preferred)

print(f"The country {country_with_most_vips_preferred['Country']} has got the most VIP customers with {country_with_most_vips_preferred['VIP+Preferred Count']} VIP and Preferred Customers")
```
